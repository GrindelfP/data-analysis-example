```python
def plot_data():
    dataFrame: DataFrame = pd.DataFrame(
        list(zip(year, pop, gdp_cap, life_exp, col)),
        columns = ['year', 'pop', 'gdp_cap', 'life_exp', 'col']
        )

    dataFrame: DataFrame = dataFrame.sort_values('pop', ascending=False) #    
    dataFrame: DataFrame = dataFrame.reset_index(drop=True) #  

    figure = plt.figure(figsize=(10, 6)) #  
    subplot = figure.add_subplot() #  
    subplot.scatter(
        x=dataFrame["gdp_cap"].tolist(),
        y=dataFrame["life_exp"].tolist(),
        s=[(dataFrame["pop"].mean())**0.5 * el for el in dataFrame["pop"].tolist()],
        marker='o',
        c=dataFrame["col"].tolist()
        ) #  
    subplot.set_xscale('log')
    subplot.xaxis.set_major_formatter(lambda x, pos: str(int(x // 100))+'k')
    
    plt.text(dataFrame.loc[0, 'gdp_cap'], dataFrame.loc[0, 'life_exp'] + (dataFrame.loc[0, 'pop'] / dataFrame['pop'].mean()) ** 0.5 + 1, "China")
    plt.text(dataFrame.loc[1, 'gdp_cap'] - 800, dataFrame.loc[1, 'life_exp'] + (dataFrame.loc[1, 'pop'] / dataFrame['pop'].mean()) ** 0.5 + 1, "India")
    
    for spine in plt.gca().spines.values(): # 
        spine.set_visible(True)
        spine.set_edgecolor('black')
        spine.set_linewidth(2)
    
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(title)
    plt.xlim([0, 100_000])
    plt.show()


plot_data()
```
